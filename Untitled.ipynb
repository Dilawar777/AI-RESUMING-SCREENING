```python
import pdfplumber
import os
import re
```

```python
def extract_text_from_pdf(pdf_path):
    """Extract all text from a PDF file."""
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text
def clean_text(text):
    """Basic text cleaning"""
    text = re.sub(r'\s+' , ' ' , text)
    text = re.sub(r'[^\w\s\.,\-\+\#\/]' , ' ' , text)
    text = text.strip()
    return text
```

```python
SKILL_KEYWORDS = {
    "Programming": ["python", "sql", "java", "c++", "r", "javascript", "scala"],
    "Data Analysis": ["pandas", "numpy", "excel", "eda", "statistical analysis",
                       "data analysis", "data cleaning", "data wrangling"],
    "Visualization": ["tableau", "power bi", "matplotlib", "seaborn", "plotly", "looker"],
    "Machine Learning": ["scikit-learn", "machine learning", "random forest",
                          "logistic regression", "deep learning", "neural network",
                          "tensorflow", "pytorch", "xgboost", "svm", "k-means",
                          "clustering", "regression", "classification"],
    "Databases": ["mysql", "postgresql", "mongodb", "sql server", "oracle",
                  "database design", "jdbc", "nosql"],
    "Tools": ["git", "github", "jupyter", "streamlit", "docker", "aws",
              "azure", "gcp", "linux", "jira"],
    "Soft Skills": ["communication", "teamwork", "leadership", "problem solving",
                     "analytical thinking", "collaboration"]
}

ALL_SKILLS = [skill for category in SKILL_KEYWORDS.values() for skill in category]
def load_text(filepath):
    with open (filepath , "r" , encoding = "utf-8") as f:
        return f.read()

def extract_skills(text , skill_list):
    """Find which skills from the list appear in the text."""
    text_lower = text.lower()
    found = []
    for skill in skill_list:
        pattern = r'/b' + re.escape(skill.lower()) + r'/b'
        if re.search(pattern , text_lower):
            found.append(skill)
    return found
def extract_skills_by_category(text, skill_dict):
    results = {}
    for category, skills in skill_dict.items():
        found = extract_skills(text, skills)
        results[category] = found
    return results
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_tfidf_score(resume_text, job_desc):
    vectorizer = TfidfVectorizer(stop_words="english", ngram_range=(1,2))
    tfidf_matrix = vectorizer.fit_transform([resume_text, job_desc])
    score = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]
    return round(score * 100, 2)

tfidf_score = calculate_tfidf_score(resume_text, SAMPLE_JOB_DESC)
print(f"Keyword match : {match_pct:.1f}%")
print(f"TF-IDF score  : {tfidf_score:.1f}%")
print(f"Final score   : {(match_pct * 0.6 + tfidf_score * 0.4):.1f}%")
```

```python
def generate_feedback(match_pct, missing_skills, matched_skills, final_score):
    print("=" * 55)
    print("RESUME SCREENING REPORT")
    print("=" * 55)
    print(f"Overall Score     : {final_score:.1f}%")
    
    if final_score >= 75:
        verdict = "STRONG MATCH — likely to pass ATS screening"
    elif final_score >= 50:
        verdict = "MODERATE MATCH — consider adding missing skills"
    else:
        verdict = "WEAK MATCH — resume needs significant improvement"
    
    print(f"Verdict           : {verdict}")
    print(f"\nSkills matched    : {len(matched_skills)}")
    print(f"Skills missing    : {len(missing_skills)}")
    
    if missing_skills:
        print(f"\nMissing keywords to add:")
        for skill in missing_skills:
            print(f"  ✗ {skill}")
    
    print(f"\nTop matched skills:")
    for skill in matched_skills[:8]:
        print(f"  ✓ {skill}")

final_score = match_pct * 0.7 + tfidf_score * 0.3
generate_feedback(match_pct, missing, matched, final_score)
```
